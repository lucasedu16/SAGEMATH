```python
var('x') # não precisa declarar x, mas recomendamos
f(x) = 2*x^2 + cos(3*x+1) - sqrt(x^2+1) # criando a função
plot(f(x), (x,-3,3)) # esboçando o gráfico
```

```python
f(x) = cos(x)
plot(f(x), (x, -pi, pi), color = 'black', fill = true, fillcolor = 'red', axes_labels=['$x$', '$y$'])
```

```python
f = plot(cos(x), (x, -2*pi, 2*pi), color = 'red', legend_label = r'$\cos(x)$')
show(f)
```

```python
g = plot(sin(x), (x, -2*pi, 2*pi), color = 'black', legend_label = r'$sin(x)$')
show(g)
```

```python
f + g
```

```python
a = plot(sin(x), (-5, -1), color = hue(0.1), linestyle = 'dashed',  ticks = [pi/2, sqrt(pi)], tick_formatter = [pi, sqrt(pi)], axes_labels = ['$x$', '$y$'])
b = plot(x^2, (-1, 2), color = (0, 0.5, 0))
c = plot(x, (2, 4), color = '#FF00FF')
show(a + b + c, aspect_ratio = 1)
```

```python
t = var('t')
v = plot(1000/(1 + 9*e^(-0.08*t)), (t, 0, 80), axes_labels = ['$t$', '$p$'], gridlines = true, title = '$Gráfico$', legend_label = '$y = x$')
u = plot(1000/(1 + 9*e^(-0.08*t)), (t, 1, 50), fill = true, fillcolor = 'black')

show(v + u)
```
